```python
import numpy as numpy
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel('B_Ordenes_04-10-2021.xlsx')
data.head(3)
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe()
```

```python
plt.figure(figsize=(10,25))
eje_x = data['PRODUCTO O SERVICIO'] 
eje_y = data['CANTIDAD PEDIDA']
plt.barh(eje_x, eje_y)
plt.ylabel('PRODUCTO O SERVICIO')
plt.xlabel('CANTIDAD PEDIDA')
plt.title('CANTIDAD PEDIDA POR PRODUCTO O SERVICIO')
plt.show()
```

```python
plt.figure(figsize=(10,10))
eje_x = data['PROVEEDOR'] 
eje_y = data['CANTIDAD PEDIDA']
plt.barh(eje_x, eje_y)
plt.ylabel('PROVEEDOR')
plt.xlabel('CANTIDAD PEDIDA')
plt.title('CANTIDAD PEDIDA POR PROVEEDOR')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(data['CANTIDAD PEDIDA'], bins=15)
plt.ylabel('FRECUENCIA')
plt.xlabel('CANTIDAD PEDIDA')
plt.title('FRECUENCIA DE CANTIDAD PEDIDA')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(data['CANTIDAD RECIBIDA'], bins=15)
plt.ylabel('FRECUENCIA')
plt.xlabel('CANTIDAD RECIBIDA')
plt.title('FRECUENCIA DE CANTIDAD RECIBIDA')
plt.show()
```
